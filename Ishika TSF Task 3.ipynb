## Bivariate analysis 

```python
categorical_sum= sample.groupby("Category").sum()
print(categorical_sum)
```

```python
categorical_sum[['Sales', 'Profit']].plot.bar()
```

Observation: Techology Products has the highest Sales and Profit, So we should encourage the sell of office supplies.
Furniture has moderate Sales but less Profit, So we should limit the sale of furniture.
Office Suplies has the least Sales but moderate Profit.

```python
subcategorical_sum= sample.groupby("Sub-Category").sum()
subcategorical_sum
```

```python
subcategorical_sum[['Sales', 'Profit']].plot(kind = 'bar', figsize = [20,8])
plt.ylabel('Total Sales/Profits',fontsize = 15)
```

Observation : From the above diagram we can see that tables have the lowest profit and Copiers have the highest profit. The sales is highest in case of phones and lowest in case of Fasteners.

```python
sample.groupby(by ='State')['Profit'].sum().sort_values(ascending = True).plot(kind = 'bar',figsize=[20,8])
plt.ylabel('Profit',fontsize = 15)
```

Observation: Maximum profit is in California and maximum loss is in Texas

```python
plt.figure(figsize=(20,10))
sns.lineplot(x='Discount',y='Profit',data=sample)
plt.title("Discount and Profit")
```

```python
plt.figure(figsize=(20,10))
sns.lineplot(x='Discount',y='Sales',data=sample)
plt.title('Discount VS Sales')
```

Observation: The graph clearly shows that if we give more discount on our products, sales increases but profit decreases.

## Conclusion 

1. When Discount > 30%, it would result in loss whereas discount<30% is making profit.
2. We should limit sales of furniture and increase that of technology and office suppliers as furniture has very less profit as compared to sales.
3. According to the sub-categories, there is huge loss on the sale of tables. So its sale should be minimized.
4. We should concentrate on the states like 'New York' and 'California' to make more profits.
